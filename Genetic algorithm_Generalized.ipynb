```python
import random as r
import numpy as n

#ALL INPUTS REQUIRED TO BE ENTERED ARE IN THE BLOCK BELOW
#---------------------------------------------------------------------------------------------------------------------------

#FUNCTION - y=((x0**2)+x1-11)**2+(x0+(x1**2)-7)**2 (variables as x[0],x[1],x[2] and so on)
#Eg y=x0**2)+(x1**2)+(x2**2) is written as(x[0]**2)+(x[1]**2)+(x[2]**2)

def function(x):
    y=(x[0]**2)+(x[1]**2)+(x[2]**2)+(x[3]**2) #WRITE YOUR FUNCTION HERE
    return y

#Initializations

k=0 #0 to find minimum of function and 1 to find maximum of function
nv=4  #number of variables
xi=0 #min value of x variables
xa=1 #max value of x variables

#These options can be left default if the user is unfamiliar with the parameters. Increase g to increase iterations.
nb=15 #number of bits per variable
np=60 #Number of points in the population
pc=0.8 #probability of crossover
pm=0.05 #probability of mutation
g=100 #Number of generations (iterations)

#---------------------------------------------------------------------------------------------------------------------------

p=[""]*np #population points
i_p=[""]*np #intermediate population

#Fitness Function - 0 if minimum and 1 if maximum
def fit_function(a1,b1):
    if(b1==0):
        c1=1/(1+a1)
    else:
        c1=a1
    return c1

#Making the r population for first iteration

for i in range(np):
    for j in range(nb*nv):
        p[i]=p[i]+str(r.randint(0,1))
        
#To find fitness function values for given population

for z in range(g):
    rn=n.zeros(np) #r points to select mating pool
    fit=n.zeros(np) #fitness function values
    pr=n.zeros(np) #probability values for points
    c_pr=n.zeros(np) #cumulative probability
    ms=n.zeros(np) #mating pool selects
    mn=n.zeros(np) #repetiton of points in mating pool
    ar=n.zeros(nv) #string to integer convert for variables
    mp=[""]*np #mating pool
    ie=[] #indexing
#Arrays can be reused for less space if required

    for i in range(np):
        for j in range(nv):
            ar[j]=xi+(xa-xi)*int(p[i][j*nb:(j+1)*nb],2)/(2**nb-1)
        fun=function(ar)
        fit[i]=fit_function(fun,k)
        ie.append(i)  #index useful for later processes

    #Reproduction

    af=(n.sum(fit))/np #Average Fitness Value
    for i in range(np):
        rn[i]=r.random() #Creating r numbers for the mating pool #CC
        pr[i]=fit[i]/(af*np) #probability for individual strings
        if i==0:
            c_pr[i]=pr[i]
        else:
            c_pr[i]=c_pr[i-1]+pr[i] #Cumulative probability for strings
    #To select mating pool
    for i in range(np):  
        for j in range(np):
            if j==1:
                if rn[i]>=0 and rn[i]<=c_pr[j]:
                    ms[i]=j
            else:
                if rn[i]>=c_pr[j-1] and rn[i]<=c_pr[j]:
                    ms[i]=j  # to select the string for which the r point lies in the cumulative pr range
    for i in range(np):
        for j in range(np):
            if ms[j]==i:
                mn[i]=mn[i]+1 #determining number of times string is repated
    c_1=0
    for i in range(np):
        for j in range(int(mn[i])):
            mp[int(c_1+j)]=p[i] #extracting the strings to form the mating pool
        c_1=c_1+mn[i] 

    # Cross-Over

    while ie!=[]:
        a=n.random.choice(ie)  #r crossing over of strings from the mating pool
        b=n.random.choice(ie)
        if(a!=b):
            if(n.random.choice([0,1],p=[1-pc,pc])==1): #Applying the probability for crossover
                site=r.randint(0,np-1) #r site for crossover
                str1=mp[a][:site]+mp[b][site:]
                str2=mp[b][:site]+mp[a][site:]
            else:
                str1=mp[a]
                str2=mp[b]
            i_p[a]=str1
            i_p[b]=str2
            ie.remove(a)
            ie.remove(b)     

    # Mutation

    for i in range(np):
        for j in range(nb*nv):
            if(n.random.choice([0,1],p=[1-pm,pm])==1): #Applying the probability for mutation
                if i_p[i][j]=='1':
                    i_p[i]=i_p[i][:j]+'0'+i_p[i][j+1:]
                else:
                    i_p[i]=i_p[i][:j]+'1'+i_p[i][j+1:]
    p=i_p
    
#Extracting the optimal point

ar=n.zeros(nv)
ar2=n.zeros(nv)
mx=0
for i in range(np):
    for j in range(nv):
        ar[j]=xi+(xa-xi)*int(p[i][j*nb:(j+1)*nb],2)/(2**nb-1)
    fun=function(ar)
    fit=fit_function(fun,k)
    if(fit>mx):
        for j in range(nv):
            ar2[j]=ar[j]
        mx=fit
if(k==0):
    print("MINIMUM POINT")
else:
    print("MAXIMUM POINT")
for i in range(nv):
    print("x",i," = ",ar2[i])
print("Function value at that point= ",function(ar2))
```
